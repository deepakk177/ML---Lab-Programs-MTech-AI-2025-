Q1. Write a Finction to calculate the  factorial of a number.

```python
def factorial(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer")

    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1

    result = 1
    for i in range(2, n + 1):
        result *= i

    return result

if __name__ == "__main__":
    test_numbers = [0, 1, 5, 10]

    print("Testing factorial function:")
    for num in test_numbers:
        result = factorial(num)
        print(f"factorial({num}) = {result}")
```

Q2. Print the first 10 numbers of the fibonacci series.

```python
def fibonacci_series(n):
    fib_series = [0, 1]

    for i in range(2, n):
        next_num = fib_series[i-1] + fib_series[i-2]
        fib_series.append(next_num)

    return fib_series

def print_fibonacci(n):
    if n <= 0:
      fib_numbers = fibonacci_series(n)

if __name__ == "__main__":
    print_fibonacci(10)

    print(f"\nFibonacci series: {fibonacci_series(10)}")
```

Q3. Check if a string is palindrome.

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]


def is_palindrome_alphanumeric(s):
    cleaned = ''.join(char.lower() for char in s if char.isalnum())
    return cleaned == cleaned[::-1]


def check_palindrome(text):
    if is_palindrome_alphanumeric(text):
        print(f"'{text}' is a palindrome!")
    else:
        print(f"'{text}' is not a palindrome.")


if __name__ == "__main__":
    print("Palindrome Checker\n")

    test_strings = [
        "racecar",
        "hello",
        "A man a plan a canal Panama",
    ]

    for string in test_strings:
        check_palindrome(string)
```

Q4. Without using the `max()`, find the largest number in a list.

```python
def find_largest(numbers):
    if not numbers:
        return None

    largest = numbers[0]

    for num in numbers[1:]:
        if num > largest:
            largest = num

    return largest

if __name__ == "__main__":
    print("Largest Number Finder")

    test_lists = [
        [45, 22, 88, 56, 92, 33],
        [-5, -2, -10, -1],
        [100],
        []
    ]

    for lst in test_lists:
        largest = find_largest(lst)
        if largest is not None:
            print(f"List: {lst}")
            print(f"Largest number: {largest}\n")

        else:
            print("List: [] (empty)")
            print("No largest number (empty list)")
```

Q5. Count the number of vovels in a string.

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0

    for char in text:
        if char in vowels:
            count += 1

    return count

if __name__ == "__main__":
    test_strings = [
        "Hello World",
        "Programming",
        "AEIOUaeiou",
    ]

    for string in test_strings:
        vowel_count = count_vowels(string)

        print(f"Text: '{string}'")
        print(f"Total vowels: {vowel_count}\n")
```

Q6. Write a program to swap the integers, without using `swap()` functiion

```python
def swap_variables(a, b):
    print(f"Before swap: a = {a}, b = {b}")
    a, b = b, a
    print(f"After swap: a = {a}, b = {b}")
    return a, b
if __name__ == "__main__":
    x, y = 10, 20
    swap_variables(x, y)
    print()

    while True:
        try:
            user_input = input("Enter two values separated by space (or 'quit' to exit): ")

            if user_input.lower() == 'quit':
                print("Goodbye!")
                break

            values = user_input.split()
            if len(values) != 2:
                print("Please enter exactly two values")
                continue

            a, b = values[0], values[1]
            swap_variables(a, b)
            print()

        except Exception as e:
            print(f"An error occurred: {e}")
```
