Q1. Write a program to predict student exam scores based on the number of hours they studied. Use linear regression to predict the scores of a student who studies for 9.5 hours.



*   Print the slope and intercept of the regression line
*   Plot regression line with dataset



```python
with open('/content/sample_data/student_scores.csv', 'r') as f:
    lines = f.readlines()

data_lines = lines[1:]

hours_studied = []
scores = []

for line in data_lines:
    parts = line.strip().split(',')
    if len(parts) == 2:
        try:
            hours_studied.append(float(parts[0]))
            scores.append(float(parts[1]))
        except ValueError:
            pass

mean_hours = sum(hours_studied) / len(hours_studied)
mean_scores = sum(scores) / len(scores)

numerator = 0
denominator = 0
for i in range(len(hours_studied)):
    numerator += (hours_studied[i] - mean_hours) * (scores[i] - mean_scores)
    denominator += (hours_studied[i] - mean_hours) ** 2

slope = numerator / denominator
intercept = mean_scores - slope * mean_hours

study_hours_to_predict = 9.5
predicted_score = intercept + slope * study_hours_to_predict

print(f"Calculated Slope: {slope}")
print(f"Calculated Intercept: {intercept}")
print(f"Predicted score for {study_hours_to_predict} hours of study: {predicted_score}")

import matplotlib.pyplot as plt

plt.scatter(hours_studied, scores, label='Original Data')

min_hours = min(hours_studied)
max_hours = max(hours_studied)
regression_x = [min_hours, max_hours]

regression_y = [intercept + slope * x for x in regression_x]

plt.plot(regression_x, regression_y, color='red', label='Regression Line')

plt.scatter(study_hours_to_predict, predicted_score, color='green', marker='X', s=100, label=f'Predicted Score ({study_hours_to_predict} hours)')

plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.title("Student Scores vs. Hours Studied with Regression Line")
plt.legend()

plt.show()
```
