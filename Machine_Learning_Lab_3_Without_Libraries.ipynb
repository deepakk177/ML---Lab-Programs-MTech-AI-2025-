Given two vectors:

x = [2, 3, 5, 7,5,6,3,1,19,11,12,16,15,5,2,1,7,2,4,5,6,12,11,2,3]

y = [1, 1, 2, 3, 2, 3, 5, 7,5,6,3,1,19,11,10,16,14,5,2,1,8,2,4,5,2]


1. Write a function minkowski_distance(x, y, p) that calculates the Minkowski distance of order p.

2. Using this function, calculate the following distances between  x and y :



*  L1 distance (p=1)
*  L2 distance (p=2)

*   L3 distance (p=3)


3. Print the results with appropriate labels.

```python
def minkowski_distance(x, y, p):

    # Minkowski distance formula: (Σ|xi - yi|^p)^(1/p)
    return sum(abs(xi - yi) ** p for xi, yi in zip(x, y)) ** (1/p)

x = [2, 3, 5, 7, 5, 6, 3, 1, 19, 11, 12, 16, 15, 5, 2, 1, 7, 2, 4, 5, 6, 12, 11, 2, 3]
y = [1, 1, 2, 3, 2, 3, 5, 7, 5, 6, 3, 1, 19, 11, 10, 16, 14, 5, 2, 1, 8, 2, 4, 5, 2]

l1_distance = minkowski_distance(x, y, 1)
l2_distance = minkowski_distance(x, y, 2)
l3_distance = minkowski_distance(x, y, 3)

print(f"L1 distance : {l1_distance}")
print(f"L2 distance : {l2_distance:.4f}")
print(f"L3 distance: {l3_distance:.4f}")
```

Part B: Linear Regression

	1.	From the Diabetes dataset, use BMI (bmi) as the independent variable (X) and the disease progression (target) as the dependent variable (y).
	2.	Print the slope (coefficient) and intercept of the regression line.
	3.	Predict the disease progression score for a patient with a BMI value of 0.05.
	4.	Plot a scatter plot of BMI vs target along with the regression line.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes

# Load dataset
diabetes = load_diabetes()
bmi = [x[2] for x in diabetes.data]  # Extract BMI feature (index 2)
disease = list(diabetes.target)

print("PART A: MINKOWSKI DISTANCE")

# Calculate distances between first two samples
x1, y1 = bmi[0], disease[0]
x2, y2 = bmi[1], disease[1]

l1 = abs(x1 - x2) + abs(y1 - y2)
l2 = ((x1 - x2)**2 + (y1 - y2)**2)**0.5
l3 = (abs(x1 - x2)**3 + abs(y1 - y2)**3)**(1/3)

print(f"L1 distance: {l1:.4f}")
print(f"L2 distance: {l2:.4f}")
print(f"L3 distance: {l3:.4f}")

print("\nPART B: LINEAR REGRESSION")

# Calculate slope and intercept manually
n = len(bmi)
sum_x = sum(bmi)
sum_y = sum(disease)
sum_xy = sum(x * y for x, y in zip(bmi, disease))
sum_x2 = sum(x * x for x in bmi)

slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
intercept = (sum_y - slope * sum_x) / n

print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")

# Predict for BMI = 0.05
bmi_test = 0.05
prediction = slope * bmi_test + intercept
print(f"Prediction for BMI = {bmi_test}: {prediction:.4f}")

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(bmi, disease, color='blue', alpha=0.6, label='Data')

# Regression line
min_bmi, max_bmi = min(bmi), max(bmi)
plt.plot([min_bmi, max_bmi],
         [slope * min_bmi + intercept, slope * max_bmi + intercept],
         color='red', linewidth=2, label='Regression line')

# Prediction point
plt.scatter([bmi_test], [prediction], color='green', s=100, marker='*',
           label=f'Prediction ({bmi_test}, {prediction:.1f})')

plt.xlabel('BMI (normalized)')
plt.ylabel('Disease Progression')
plt.title('BMI vs Disease Progression')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nEquation: y = {slope:.4f}x + {intercept:.4f}")
```
